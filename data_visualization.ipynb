```python
# Import data analysis libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Import data from csv file
data = pd.read_csv('tarifas.csv')
```

```python
# Display the rows of the data
data.head(20)
```

```python
# Display the information of the data
data.info()
```

```python
# Visualize the Punta by company in horizontal bar plot
sns.set_style(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.barplot(x='Punta', y='Empresa', data=data, palette='viridis')
plt.title('Punta by Company')
plt.xlabel('Punta')
plt.ylabel('Company')
plt.show()
```

```python
# Visualize the Valle by company in horizontal bar plot
sns.set_style(style='darkgrid')
plt.figure(figsize=(10, 6))
sns.barplot(x='Valle', y='Empresa', data=data, palette='viridis')
plt.title('Valle by Company')
plt.xlabel('Valle')
plt.ylabel('Company')
plt.show()
```

```python
# Dashboard with all the plots
sns.set_style(style='darkgrid')

# Crear una figura con un tablero de 3x2
fig, axes = plt.subplots(3, 2, figsize=(15, 15))

# Agregar un título al dashboard
fig.suptitle('Comparativa de Tarifas y Empresas', fontsize=16)

# Gráfico 1: Punta por Empresa
sns.barplot(x='Punta', y='Empresa', data=data, palette='viridis', ax=axes[0, 0])
axes[0, 0].set_title('Punta por Empresa')
axes[0, 0].set_xlabel('Punta')
axes[0, 0].set_ylabel('Empresa')

# Gráfico 2: Punta por Tarifa
sns.barplot(x='Punta', y='Tarifa', data=data, palette='viridis', ax=axes[0, 1])
axes[0, 1].set_title('Punta por Tarifa')
axes[0, 1].set_xlabel('Punta')
axes[0, 1].set_ylabel('Tarifa')

# Gráfico 3: Llano por Empresa
sns.barplot(x='Llano', y='Empresa', data=data, palette='viridis', ax=axes[1, 0])
axes[1, 0].set_title('Llano por Empresa')
axes[1, 0].set_xlabel('Llano')
axes[1, 0].set_ylabel('Empresa')

# Gráfico 4: Llano por Tarifa
sns.barplot(x='Llano', y='Tarifa', data=data, palette='viridis', ax=axes[1, 1])
axes[1, 1].set_title('Llano por Tarifa')
axes[1, 1].set_xlabel('Llano')
axes[1, 1].set_ylabel('Tarifa')

# Gráfico 5: Valle por Empresa
sns.barplot(x='Valle', y='Empresa', data=data, palette='viridis', ax=axes[2, 0])
axes[2, 0].set_title('Valle por Empresa')
axes[2, 0].set_xlabel('Valle')
axes[2, 0].set_ylabel('Empresa')

# Gráfico 6: Valle por Tarifa
sns.barplot(x='Valle', y='Tarifa', data=data, palette='viridis', ax=axes[2, 1])
axes[2, 1].set_title('Valle por Tarifa')
axes[2, 1].set_xlabel('Valle')
axes[2, 1].set_ylabel('Tarifa')

# Ajustar el layout
plt.tight_layout(rect=[0, 0, 1, 0.95]) # Ajustar para dejar espacio para el título
plt.show()
```

```python
# Load the data from the csv file
data_solar = pd.read_csv('datos_solares2.csv')

# Display the rows of the data
data_solar.head(10)
```

```python
# Display the information of the data
data_solar.info()
```

```python
# Filtrar las diez ciudades con más horas de sol
top_cities = data_solar.nlargest(10, 'Horas de sol')

# Visualizar las diez ciudades con más horas de sol
sns.set_style(style='darkgrid')
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Horas de sol', y='Ciudad', data=top_cities, palette='viridis')
plt.title('Top Ten Cities with the Highest Hours of Sun')
plt.xlabel('Hours of Sun')
plt.ylabel('City')

# Añadir etiquetas con los valores dentro de las barras
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

plt.show()
```

```python
# Filtrar las diez ciudades con menos horas de sol
bottom_cities = data_solar.nsmallest(10, 'Horas de sol')

# Visualizar las diez ciudades con menos horas de sol
sns.set_style(style='darkgrid')
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Horas de sol', y='Ciudad', data=bottom_cities, palette='viridis')
plt.title('Top Ten Cities with the Least Hours of Sun')
plt.xlabel('Hours of Sun')
plt.ylabel('City')

# Añadir etiquetas con los valores dentro de las barras
for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')

plt.show()
```
